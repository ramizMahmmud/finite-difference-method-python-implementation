```python
import numpy as np
import matplotlib.pyplot as plt

## Numerical Solution
def solver(I,w,dt,T):
    """
    Solve u" + w**2*u = 0 for t in (0,T], u(0) = I and u'(0) = 0
    by central difference method
    """
    dt = float(dt)
    Nt = int(round(T/dt))
    u = np.zeros(Nt+1)
    t = np.linspace(0,Nt*dt,Nt+1)
    u[0] = I
    u[1] = u[0] - 0.5 * dt**2 * w**2* u[0]
    for n in range(1, Nt):
        u[n+1] = 2*u[n] - u[n-1] - dt**2 * w**2 * u[n]
    return u, t 


## Exact Solution
def u_exact(t, I, w):
    return I*np.cos(w*t)

## Visualization
def visulaize(u,t):
    plt.plot(t,u)


# Example
I = 1
w = 2* np.pi
num_period = 5
p = 2 * np.pi / w # one period
dt = 0.05
T = num_period * p
u, t = solver(I, w, dt, T)
v = u_exact(t, I, w)

visulaize(u,t)



```

```python
visulaize(v,t)
```
